```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
import math

def convOI(Cin,Cout,Wout,Hout,Fw,Fh,B):
    return 2*Cout*(Cin*Fw*Fh)*Wout*Hout*B/(4.0*(2*((Cin*Fw*Fh+1)*Cout + Cout*Wout*Hout*B + Cin*Fw*Fh*Wout*Hout*B)));

def FCOI(Bin,Bout,B):
    return 2*(Bin)*Bout*B/(4.0*(2*((Bin+1)*Bout + (Bin+Bout)*B)));

def MaxPoolOI(Cout,Wout,Hout,Fw,Fh):
    return 2*Cout*Wout*Hout*Fw*Fh/(4.0*(2*Cout*Wout*Hout));

def ReluOI(Cout,Wout,Hout):
    return (2*Cout*Wout*Hout)/(4.0*(2*Cout*Wout*Hout));



def convFLOPs_Model(Cin,Cout,Wout,Hout,Fw,Fh,B):
    return 2*Cout*(Cin*Fw*Fh)*Wout*Hout*B;
def convMemory_Model(Cin,Cout,Wout,Hout,Fw,Fh,B):
    return (4*2*((Cin*Fw*Fh+1)*Cout + Cout*Wout*Hout*B));

def FCFLOPs_Model(Bin,Bout,B):
    return 2*(Bin)*Bout*B;
def FCMemory_Model(Bin,Bout,B):
    return (4*2*((Bin+1)*Bout + Bout*B));

def MaxPoolFLOPs_Model(Cout,Wout,Hout,Fw,Fh,B):
    return 2*Cout*Wout*Hout*Fw*Fh*B;
def MaxPoolMemory_Model(Cout,Wout,Hout,Fw,Fh,B):
    return (4*(2*Cout*Wout*Hout*B));

def ReluFLOPs_Model(Cout,Wout,Hout,B):
    return 2*Cout*Wout*Hout*B;
def ReluMemory_Model(Cout,Wout,Hout,B):
    return (4*(2*Cout*Wout*Hout*B));



def convMemory_reCompute(Cin,Cout,Wout,Hout,Fw,Fh):
    return 4*2*(Cin*Fw*Fh+1)*Cout;
def FCMemory_reCompute(Bin,Bout):
    return 4*2*((Bin+1)*Bout );
def MaxPoolMemory_reCompute(Cout,Wout,Hout,Fw,Fh):
    return 4*(2*Cout*Wout*Hout);
def ReluMemory_reCompute(Cout,Wout,Hout):
    return 4*(2*Cout*Wout*Hout);
```

```python
#compute OI of VGG16 on the given Input data size and batch size
#return a list of OI distribution on each layer in VGG16
def VGG16OI(Input,B):
    OI=[];
    for block in range(0,7):
        tempOI=0;
        if block==0 :
            tempOI=convOI(3,64,Input,Input,3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(64,Input,Input);
            OI.append([tempOI]);
            tempOI=convOI(64,64,Input,Input,3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(64,Input,Input);
            OI.append([tempOI]);
            tempOI=MaxPoolOI(64,math.ceil(Input/2),math.ceil(Input/2),2,2);
            OI.append([tempOI]);
        elif block==1 :
            tempOI=convOI(64,128,math.ceil(Input/2),math.ceil(Input/2),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(128,math.ceil(Input/2),math.ceil(Input/2));
            OI.append([tempOI]);
            tempOI=convOI(128,128,math.ceil(Input/2),math.ceil(Input/2),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(128,math.ceil(Input/2),math.ceil(Input/2));
            OI.append([tempOI]);
            tempOI=MaxPoolOI(128,math.ceil(Input/4),math.ceil(Input/4),2,2);
            OI.append([tempOI]);
        elif block==2 :
            tempOI=convOI(128,256,math.ceil(Input/4),math.ceil(Input/4),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(256,math.ceil(Input/4),math.ceil(Input/4));
            OI.append([tempOI]);
            tempOI=convOI(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(256,math.ceil(Input/4),math.ceil(Input/4));
            OI.append([tempOI]);
            tempOI=convOI(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(256,math.ceil(Input/4),math.ceil(Input/4));
            OI.append([tempOI]);
            tempOI=MaxPoolOI(256,math.ceil(Input/8),math.ceil(Input/8),2,2);
            OI.append([tempOI]);
        elif block==3 :
            tempOI=convOI(256,512,math.ceil(Input/8),math.ceil(Input/8),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/8),math.ceil(Input/8));
            OI.append([tempOI]);
            tempOI=convOI(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/8),math.ceil(Input/8));
            OI.append([tempOI]);
            tempOI=convOI(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/8),math.ceil(Input/8));
            OI.append([tempOI]);
            tempOI=MaxPoolOI(512,math.ceil(Input/16),math.ceil(Input/16),2,2);
            OI.append([tempOI]);
        elif block==4 :
            tempOI=convOI(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/16),math.ceil(Input/16));
            OI.append([tempOI]);
            tempOI=convOI(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/16),math.ceil(Input/16));
            OI.append([tempOI]);
            tempOI=convOI(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,B);
            OI.append([tempOI]);
            tempOI=ReluOI(512,math.ceil(Input/16),math.ceil(Input/16));
            OI.append([tempOI]);
            tempOI=MaxPoolOI(512,math.ceil(Input/32),math.ceil(Input/32),2,2);
            OI.append([tempOI]);
        elif block==5 :
            tempOI=FCOI(512*math.ceil(Input/32)*math.ceil(Input/32),4096,B);
            OI.append([tempOI]);
            tempOI=ReluOI(4096,1,1);
            OI.append([tempOI]);
        elif block==6 :
            tempOI=FCOI(4096,4096,B);
            OI.append([tempOI]);
            tempOI=ReluOI(4096,1,1);
            OI.append([tempOI]);
        elif block==7 :
            tempOI=FCOI(4096,1000,B);
            OI.append([tempOI]);
    return OI;
```

```python
#return the total OI in VGG16 with given Input data size
def VGG16OI_Model(Input):
    FLOPs=0;
    Memory=0;
    for block in range(0,7):
        tempFLOPs=0;
        tempMemory=0;
        if block==0 :
            tempFLOPs=convFLOPs_Model(3,64,Input,Input,3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(64,Input,Input,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(64,64,Input,Input,3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(64,Input,Input,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=MaxPoolFLOPs_Model(64,math.ceil(Input/2),math.ceil(Input/2),2,2,1);
            FLOPs+=tempFLOPs;
            tempMemory=convMemory_Model(3,64,Input,Input,3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(64,Input,Input,1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(64,64,Input,Input,3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(64,Input,Input,1);
            Memory+=tempMemory;
            tempMemory=MaxPoolMemory_Model(64,math.ceil(Input/2),math.ceil(Input/2),2,2,1);
            Memory+=tempMemory;
        elif block==1 :
            tempFLOPs=convFLOPs_Model(64,128,math.ceil(Input/2),math.ceil(Input/2),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(128,math.ceil(Input/2),math.ceil(Input/2),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(128,128,math.ceil(Input/2),math.ceil(Input/2),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(128,math.ceil(Input/2),math.ceil(Input/2),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=MaxPoolFLOPs_Model(128,math.ceil(Input/4),math.ceil(Input/4),2,2,1);
            FLOPs+=tempFLOPs;
            tempMemory=convMemory_Model(64,128,math.ceil(Input/2),math.ceil(Input/2),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(128,math.ceil(Input/2),math.ceil(Input/2),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(128,128,math.ceil(Input/2),math.ceil(Input/2),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(128,math.ceil(Input/2),math.ceil(Input/2),1);
            Memory+=tempMemory;
            tempMemory=MaxPoolMemory_Model(128,math.ceil(Input/4),math.ceil(Input/4),2,2,1);
            Memory+=tempMemory;
        elif block==2 :
            tempFLOPs=convFLOPs_Model(128,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=MaxPoolFLOPs_Model(256,math.ceil(Input/8),math.ceil(Input/8),2,2,1);
            FLOPs+=tempFLOPs;
            tempMemory=convMemory_Model(128,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(256,256,math.ceil(Input/4),math.ceil(Input/4),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(256,math.ceil(Input/4),math.ceil(Input/4),1);
            Memory+=tempMemory;
            tempMemory=MaxPoolMemory_Model(256,math.ceil(Input/8),math.ceil(Input/8),2,2,1);
            Memory+=tempMemory;
        elif block==3 :
            tempFLOPs=convFLOPs_Model(256,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=MaxPoolFLOPs_Model(512,math.ceil(Input/16),math.ceil(Input/16),2,2,1);
            FLOPs+=tempFLOPs;
            tempMemory=convMemory_Model(256,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(512,512,math.ceil(Input/8),math.ceil(Input/8),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/8),math.ceil(Input/8),1);
            Memory+=tempMemory;
            tempMemory=MaxPoolMemory_Model(512,math.ceil(Input/16),math.ceil(Input/16),2,2,1);
            Memory+=tempMemory;
        elif block==4 :
            tempFLOPs=convFLOPs_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=convFLOPs_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            FLOPs+=tempFLOPs;
            tempFLOPs=MaxPoolFLOPs_Model(512,math.ceil(Input/32),math.ceil(Input/32),2,2,1);
            FLOPs+=tempFLOPs;
            tempMemory=convMemory_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            Memory+=tempMemory;
            tempMemory=convMemory_Model(512,512,math.ceil(Input/16),math.ceil(Input/16),3,3,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(512,math.ceil(Input/16),math.ceil(Input/16),1);
            Memory+=tempMemory;
            tempMemory=MaxPoolMemory_Model(512,math.ceil(Input/32),math.ceil(Input/32),2,2,1);
            Memory+=tempMemory;
        elif block==5 :
            tempFLOPs=FCFLOPs_Model(512*math.ceil(Input/32)*math.ceil(Input/32),4096,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(4096,1,1,1);
            FLOPs+=tempFLOPs;
            tempMemory=FCMemory_Model(512*math.ceil(Input/32)*math.ceil(Input/32),4096,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(4096,1,1,1);
            Memory+=tempMemory;
        elif block==6 :
            tempFLOPs=FCFLOPs_Model(4096,4096,1);
            FLOPs+=tempFLOPs;
            tempFLOPs=ReluFLOPs_Model(4096,1,1,1);
            FLOPs+=tempFLOPs;
            tempMemory=FCMemory_Model(4096,4096,1);
            Memory+=tempMemory;
            tempMemory=ReluMemory_Model(4096,1,1,1);
            Memory+=tempMemory;
        elif block==7 :
            tempFLOPs=FCFLOPs_Model(4096,1000,1);
            FLOPs+=tempFLOPs;
            tempMemory=FCMemory_Model(4096,1000,1);
            Memory+=tempMemory;
    return FLOPs/Memory;
```

```python
#tool to count computations in forward popagation using pytorch
#return total FLOPs value in a fp process with the given model
import torch
import torchvision

import torch.nn as nn
from torch.autograd import Variable
import torchvision.models as models

import numpy as np

def print_model_parm_flops(Model):

    prods = {}
    def save_hook(name):
        def hook_per(self, input, output):
            prods[name] = np.prod(input[0].shape)
        return hook_per

    list_1=[]
    def simple_hook(self, input, output):
        list_1.append(np.prod(input[0].shape))
    list_2={}
    def simple_hook2(self, input, output):
        list_2['names'] = np.prod(input[0].shape)


    multiply_adds = False
    list_conv=[]
    def conv_hook(self, input, output):
        batch_size, input_channels, input_height, input_width = input[0].size()
        output_channels, output_height, output_width = output[0].size()

        kernel_ops = self.kernel_size[0] * self.kernel_size[1] * (self.in_channels / self.groups) * (2 if multiply_adds else 1)
        bias_ops = 1 if self.bias is not None else 0

        params = output_channels * (kernel_ops + bias_ops)
        flops = batch_size * params * output_height * output_width

        list_conv.append(flops)


    list_linear=[] 
    def linear_hook(self, input, output):
        batch_size = input[0].size(0) if input[0].dim() == 2 else 1

        weight_ops = self.weight.nelement() * (2 if multiply_adds else 1)
        bias_ops = self.bias.nelement()

        flops = batch_size * (weight_ops + bias_ops)
        list_linear.append(flops)

    list_bn=[] 
    def bn_hook(self, input, output):
        list_bn.append(input[0].nelement())

    list_relu=[] 
    def relu_hook(self, input, output):
        list_relu.append(input[0].nelement())

    list_pooling=[]
    def pooling_hook(self, input, output):
        batch_size, input_channels, input_height, input_width = input[0].size()
        output_channels, output_height, output_width = output[0].size()

        kernel_ops = self.kernel_size * self.kernel_size
        bias_ops = 0
        params = output_channels * (kernel_ops + bias_ops)
        flops = batch_size * params * output_height * output_width

        list_pooling.append(flops)


            
    def foo(net):
        childrens = list(net.children())
        if not childrens:
            if isinstance(net, torch.nn.Conv2d):
                net.register_forward_hook(conv_hook)
            if isinstance(net, torch.nn.Linear):
                net.register_forward_hook(linear_hook)
            if isinstance(net, torch.nn.BatchNorm2d):
                net.register_forward_hook(bn_hook)
            if isinstance(net, torch.nn.ReLU):
                net.register_forward_hook(relu_hook)
            if isinstance(net, torch.nn.MaxPool2d) or isinstance(net, torch.nn.AvgPool2d):
                net.register_forward_hook(pooling_hook)
            return
        for c in childrens:
                foo(c)

    foo(Model)
    input = Variable(torch.rand(3,224,224).unsqueeze(0), requires_grad = True)
    out = Model(input)


    total_flops = (sum(list_conv) + sum(list_linear) + sum(list_bn) + sum(list_relu) + sum(list_pooling))
    
    print('  Number of FLOPs: %.2fG' % (total_flops / 1e9))
```

As we've talked before, the computation and memory cost for each layer in CNN are:

Computation(FLOPs):
   
   Convolutional Layer:
    $2C_{out}(2C_{in}F_wF_h+1)W_{out}H_{out}B$
    
   Relu Layer:
    $2C_{out}W_{out}H_{out}B$ (Comp)
    
   Max Pooling Layer:
    $2C_{out}W_{out}H_{out}F_wF_h$
    
   Fully Connected Layer:
    $2(B_{in}+1)B_{out}$
    
Memory(Byte)(assume we are using float32):

   Convolutional Layer:
    $4\times2[(C_{in}F_wF_h+1)C_{out}+C_{out}W_{out}H_{out}B+C_{in}F_wF_hW_{out}H_{out}B]$
   
   Relu Layer:
    $4\times2C_{out}W_{out}H_{out}B$
   
   Max Pooling Layer:
    $4\times2C_{out}W_{out}H_{out}B$
   
   Fully Connected Layer:
    $4\times2[(B_{in}+1)B_{out}+(B_{in}+B_{out})B]$
    
Operational Intensity = Computation/Memory

However, when we're counting the whole memory cost for a model, we don't count input feature maps, which is $C_{in}F_wF_hW_{out}H_{out}B$ in convolutional layers and $B_{in}B$ in fully connected layers, to avoid redundant counts.
